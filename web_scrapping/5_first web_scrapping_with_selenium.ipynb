```python
from selenium import webdriver
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
products=[] #List to store name of the product
prices=[] #List to store price of the product
ratings=[] #List to store rating of the product
driver.get("https://www.flipkart.com/laptops-store")
```

```python
content = driver.page_source
```

```python
content
```

```python
soup = BeautifulSoup(content)
```

```python
soup
```

```python
for div in soup.findAll('div',href=False, attrs={'class':'_4ddWXP _3BCh3_'}):
    name=div.find('a', attrs={'class':'s1Q9rs'})
    price=div.find('div', attrs={'class':'_30jeq3'})
    rating=div.find('div', attrs={'class':'_3LWZlK'})
    products.append(name.attrs['title'])
    prices.append(price.text)
    ratings.append(rating.text)
```

```python
df = pd.DataFrame({'Product Name':products,'Price':prices,'Rating':ratings})
df.to_csv('products.csv', index=False, encoding='utf-8')
df.to_excel('products.xlsx', index=False, encoding='utf-8')
```
